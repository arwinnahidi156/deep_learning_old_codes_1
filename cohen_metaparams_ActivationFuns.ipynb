```python
import torch
import matplotlib.pyplot as plt
```

```python
x=torch.linspace(-3,3,101)
actfun=getattr(torch,'relu')
def NNoutputx(actfun):
    actfun=getattr(torch,actfun)
    return actfun(x)
```

```python
activation_funs=['relu','sigmoid','tanh']
for actfun in activation_funs:
    plt.plot(x,NNoutputx(actfun),label=actfun)
dashlinecol=[0.7,0.7,0.7]
plt.plot(x[[0,-1]],[0,0],'--',color=dashlinecol)
plt.plot(x[[0,-1]],[1,1],'--',color=dashlinecol)
plt.plot([0,0],[-1,3],'--',color=dashlinecol)

plt.legend()

```

```python
x=torch.linspace(-3,3,101)
actfun=getattr(torch,'relu')
def NNoutputx(actfun):
    actfun=getattr(torch.nn,actfun)
    return actfun()
```

```python
activation_funs=['ReLU6','Hardshrink','LeakyReLU']
for actfun in activation_funs:
    plt.plot(x,NNoutputx(actfun)(x),label=actfun)
dashlinecol=[0.7,0.7,0.7]
plt.plot(x[[0,-1]],[0,0],'--',color=dashlinecol)
plt.plot(x[[0,-1]],[1,1],'--',color=dashlinecol)
plt.plot([0,0],[-1,3],'--',color=dashlinecol)

plt.legend()

```

```python
x=torch.linspace(-3,10,401)
relu6=torch.nn.ReLU6()
plt.plot(x,relu6(x))
```

```python
x=torch.linspace(-3,3,21)
y1=torch.relu(x)
f=torch.nn.ReLU()
y2=f(x)
plt.plot(x,y1,'--',label='torch.relu')
plt.plot(x,y2,'rs',label='torch.nn.ReLU')
plt.legend()
```
